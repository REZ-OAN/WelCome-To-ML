```python
import pandas as pd
import requests as req

df = pd.DataFrame()
for i in range(1,401) :
    res = req.get('https://api.themoviedb.org/3/movie/top_rated?api_key=8265bd1679663a7ea12ac168da84d2e8&language=en-US&page={}'.format(i))
    ## creating data frame
    data = res.json()['results']
    temp_Df = pd.DataFrame(data)[['id','title','overview','release_date','popularity','vote_average','vote_count']]
    df = df.append(temp_Df,ignore_index=True)
  

```
